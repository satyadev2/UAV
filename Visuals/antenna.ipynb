```python
import numpy as np
import matplotlib.pyplot as plt

def plot_omnidirectional_pattern():
    angles = np.linspace(0, 2 * np.pi, 360)
    r = np.ones_like(angles)  # Constant radius for all angles

    plt.figure(figsize=(4, 4))
    plt.polar(angles, r, label='Omnidirectional Antenna')
    plt.title('Omnidirectional Antenna Pattern')
    plt.legend()
    plt.show()

plot_omnidirectional_pattern()
```

```python
def plot_beamforming_pattern(beamwidth=20, gain=10):
    angles = np.linspace(0, 2 * np.pi, 360)
    # Gaussian-like beam pattern for simplicity
    pattern = gain * np.exp(-((angles - np.pi) ** 2) / (2 * (beamwidth * np.pi / 180) ** 2))
    pattern = np.roll(pattern, 180)  # Align the main lobe to 0 degrees

    plt.figure(figsize=(4,4))
    plt.polar(angles, pattern, label=f'Beamforming Antenna (Beamwidth={beamwidth})')
    plt.title('Beamforming Antenna Pattern')
    plt.legend()
    plt.show()

plot_beamforming_pattern(beamwidth=20, gain=10)
```
